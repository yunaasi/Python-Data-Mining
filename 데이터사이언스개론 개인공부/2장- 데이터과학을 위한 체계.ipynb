## 1. 데이터과학을 위한 기본 지식

### 1)데이터의 유형

    ✅ 정형화에 따른 분류
        (1) 정형 : 정리되어 있어 비교적 분석하기 쉽고, 단순한 형태의 데이터로 고정된 필드에 저장된 데이터, 스프레드 시트 등의 값이 있다.
        ex. 고객 신상 데이터, 매출 데이터, 재고 데이터 등

        (2) 반정형 : 고정된 필드에 저장되어 있지 않은 정형 데이터로, 레코드 및 필드와 같은 구조를 갖기 위한 태그나 인덱스 등을 포함한다.
        ex. XML, HTML, JSON, 로그 형태 등

        (3) 비정형 : 정리되어 있지 않아 분석하기 어렵고 복잡한 형태의 데이터로 고정된 필드에 저장되어 있지 않다.
        ex. 이미지, 동영상 음성 데이터, 이메일 등 

    ✅ 시간에 따른 분류
        (1) 시계열 데이터(time series data) : 시간에 따라 관측된 데이터를 말한다. 각 관측지들이 서로 독립적이지 않는다.
        ex. 일별 온도, 연도별 벼 수확량 , 주가 등

        (2) 횡단면 데이터(cross-sectional-data) 동 시간대에 얻은 데이터를 말한다. 보통 관측지 간의 독립성이 어느정도 보장된다.
        ex. 도시별 인구수, 점포별 매출 현황 등

        (3) 패널 데이터(panel data): 횡단면 데이터를 시간에 따라 나열한 데이터로 시계열 데이터 및 횡단면 데이터가 동시 포함됨 
        ex. 연도별 점포 매출 현황, 시도별 월별 세수 현황 등이 있다.

### 2) 모집단과 평균 개념

    ✅ 모집단
        - 관심의 대상이 되는 집단으로 모집단의 특성은 해당 모집단으로 부터 추출된 표본을 이용해 추정한다. 
        - 모집단과 관련한 통계량 population : 모평균, 모분산, 모표준편차 
        - 표본과 관련한 통계량 sample: 표본평균, 표본 분산, 표본 표준 편차 
        https://stonefree.tistory.com/48

### 3) 척도와 변수 
    - 데이터는 측정 척도(scale of measurement)에 따라 명목, 순위, 등간, 비율 변수로 구분한다.
    ✅ 범주형(질적) 변수(비계량적): 명목, 순위 변수  => 로지스틱 회의분석, 선형 판별 분석
        -명목 변수 nominal variable : 서로 다름 외에 어떤 정보도 포함하고 있지 않은 데이터로 <각 변수간 연산이 불가능>하고, 명목 변수의 범주가 2개인 경우, 이항형이라고 한다
        -순위 변수 ordinal variable : 명목형 데이터 보다 크고 작음을 나타내는 정보가 추가로 포함되어 있는 데이터로 수치화해 계산하는 것이 가능하다.
    ✅ 수치형 변수 : 등간, 비율 변수  => 다중 선형 회귀분석, 분산 분석
        -등간 변수 interval variable : 간격에 대한 정보가 포함되어 있는 변수 , 덧셈과 뺄셈이 가능하지만 대부분 구분 없이 수치형 변수로 일괄 이용한다.
        -비율 변수 ration variable: 일반적으로 사용하는 수치형 변수에 해당한다.
        https://quickdata.tistory.com/58
    

    - 설명 변수와 반응 변수
    ✅ [ID] 변수 : 개별 관측지의 구분을 목적으로 하는 세분화 변수로 key라고도 하며, 데이터베이스 관리 시스템의 key와 유사한 개념이다
    ✅ 반응 변수 : 예측 또는 설명의 목적에 해당하는 변수로 <종속 변수>라고도 한다. 기계학습과 인공지능에서 label이라고 불림
    ✅ 설명 변수 : 반응 변수에 영향을 줄 먻으로 예상되는 변수 <독립변수>라도고 하며, 기계학습 및 인공지능 영역에서 feature 이라고 불림


## 2. 데이터 과학의 프레임워크

### 1) 데이터 과학 모형 및 평가 
    ✅ 데이터 과학 모형 
        - 데이터 과학에서 모형은 어떤 원인 변수들에 의해 결과 변수를 설명 또는 예측하기 위해 만들어졌다.
        - 주어진 변수를 이용해 목표하는 변수를 근사적으로 예측 및 분류하는 것 
        - 모형을 만드는 과정 = 모형 적합 or 모형 학습 
        y= f(x) + Error
    
    ✅ 모형 적합도에 대한 평가
        - 오차가 작을수록 좋은 모형이고, 모형 혜측 결과와 실측치 간의 차이를 이용해 오차를 측정할 수 있다.
        - 오차 제곱합(sum of squared error;SSE)는 실측치와 예측치 차이의 제곱합으로 모형의 정확도를 측정한다.
        - 관측치 1개당 평균적인 SSE를 나타내는 평균제곱오차를 주로 사용한다.
    
    ✅ 편향과 분산간 상충 관계  -> 데이터 오차의 세가지 유형
        (1) 분산 variance: 학습용 데이터를 잘 예측하도록 만들어진 모형이 평가용 데이터를 예측하면 발생한 데이터 간의 차이로 발생하는 오차 
        (2) 편향 vias : 모형이 너무 단순해서 예측을 제대로 하지 못하는 경우 발생하는 오차
        (3) 축소 불가능한 오차 irreducible error : 반응 변수에 영향을 주는 변수 중 일부가 수집되지 않았거나, 분석에 이용된 데이터가 갖고 있지 않은 다른 정보에 의해 발생한 오차 
                                            : 이미 데이터 자체에 내포되어 있어 이미 줄일 수 없는 오차를 의미함


### 2) 분석 목적에 따른 프레임워크
    ✅ 예측과 추론 
    ✅ 예측이 주 목적인 경우
    ✅ 추론이 주 목적인 경우
        - 설명 변수와 반응 변수(독립변수와 종속 변수)의 관계를 파악하는 것이 중요 
        - 회귀모형, 의사결정 나무 등을 활용
    ✅ 예측 정확도와 모델 해석력 간 상충 관계 
    ✅ 탐색적 데이터 분석과 확증적 데이터 분석
        - 탐색적 데이터 분석 : 데이터의 주요한 특성을 분석하기 위해 데이터를 요약하고 시각화 하는 방법을 의미
        - 확증적 데이터 분석 : 엄격하고 체계적인 방법으로 가설을 검증하는 방법을 의미

### 3) 이론적 배경에 따른 프레임워크
    ✅ 모수적 방법과 비모수적 방법
    (1) 모수적 방법
        : 표본으로부터 모집단의 특성인 모수를 추정할 때 모집단의 분포를 가정하는 방법으로, 정규분포를 가정하는 경우가 대부분이다.
    (2) 비모수적 방법
        : 모집단의 모수 자체에 관심을 두지 않은 순수 비모수적 방법 vs 검정 통계량 분포와 모집단의 분포가 서로 무관한 분포 무관 방법 으로 나뉜다.
        : 모집단의 정규성 가정이 어렵고 동시에 표본수가 작은 경우 이용함
        : 부호, 순위 혹은 순위에 기초한 점수를 이용한다.

### 4) 변수 유형에 따른 프레임워크
    ✅ 지도학습과 비지도 학습
        (1) 지도학습 : 반응 변수가 있는 경우에 해당한다.
        (2) 비지도 학습 : 반응 변수가 없는 경우, 모든 변수가 반응 변수인 경우 해당

    ✅ 회귀와 분류
        (1) 회귀 regression : 연속형 수치 예측
            - 양적 반응 변수 사용 
            - 무한 예측 범위
            - 분석 방법론 : 선형 회귀분석, 분산 순석
        
        (2) 분류 classification
            - 범주형 분류
            - 질적 반응 변수 사용
            - 유한하고 주어진 분류 유형수를 예측 범위
            - 분석 방법론 : 로지스틱 회귀, 판별 분석
    
    ✅ 그룹화와 차원 축소
        (1) 그룹화 grouping : 관측치(행,row)를 묶는 방법
            - 데이터에 기반해 그룹을 찾는 방법
            - 대표 방법론 : 군집 분석, 연관성 분석
        
        (2) 차원 축소 dimensionality reduction
            - 변수(열, column)을 줄이는 방법
            - 대표성을 갖는 변수를 찾거나 만드는 방법
            - 분석 방법론 : 변수 선택, 변수 결합
